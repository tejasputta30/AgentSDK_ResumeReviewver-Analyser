```python
import gradio as gr

def ui_function(resume_text):
    try:
        result = run_agent(resume_text)
        return result
    except Exception as e:
        return f"⚠️ Error: {str(e)}"


with gr.Blocks(title="AI Resume Reviewer") as demo:

    # 🔷 Title Section
    gr.Markdown("""
    # 📄 AI Resume Reviewer & Improver
    Get instant feedback and improve your resume using AI 🚀
    """)

    # 🔷 Input Box
    resume_input = gr.Textbox(
        label="📥 Paste Your Resume",
        placeholder="Enter your resume details (skills, education, projects, etc.)...",
        lines=8
    )

    # 🔷 Buttons
    with gr.Row():
        submit_btn = gr.Button("🚀 Analyze Resume", variant="primary")
        clear_btn = gr.Button("🗑️ Clear")

    # 🔷 Output Box
    output_box = gr.Textbox(
        label="📊 AI Feedback & Improved Resume",
        lines=15
    )

    # 🔷 Example Button
    example_btn = gr.Button("💡 Load Example")

    def load_example():
        return "I am a computer engineering student. I know Python and C++. I have basic knowledge but no projects."

    example_btn.click(load_example, None, resume_input)

    # 🔷 Button Actions
    submit_btn.click(ui_function, inputs=resume_input, outputs=output_box)
    clear_btn.click(lambda: ("", ""), None, [resume_input, output_box])

    # 🔷 Footer
    gr.Markdown("""
    ---
    ⚡ Powered by OpenAI
    💡 Tip: Add **projects, achievements, and metrics** for better results
    """)

demo.launch(share=True)
```
